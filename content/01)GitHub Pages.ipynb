## HTTP
***
HTTP [HyperText Transfer Protocol](https://developer.mozilla.org/en-US/docs/Web/HTTP/Overview) is the main protocol used for getting images, videos, ads etc or any data exchange on the web, HTTP follow a client server protocol which means the client sends a request, client then wait for a response from the server, the server sends the client back the request mostly with a status code of 200 for success 

HTTP is stateless which mean the server should not retain any state between requests, this is known as a REST [REpresentational State Transfer](https://restfulapi.net/) compliant systems which is how to set apart the client and the server


## Example Request Response
[A typical HTTP session](https://developer.mozilla.org/en-US/docs/Web/HTTP/Session)
***


Certain status codes in HTTP mean different things.

* 1xx informational response – the request was received, and the process is carrying on.
* 2xx successful – the request was successful.
* 3xx redirection – further action needs to be taken i.e the request needs to be redirected to another site possibly, in order to complete the request
* 4xx client error – the request contains bad syntax or cannot be fulfilled
* 5xx server error – the server failed to fulfil an apparently valid request

## Static vs Dynamic Websites
***

* Static means the web application stays the same the button clicks etc will not change the website permanently, content is not generated from a backend database like mongo or sql


* Website will be interative although not as widely interative as a dynamic website.


* Dynamic means the server gets a request which changes the database with user interaction such as a user creating a post in facebook


* GitHub pages is only used for static content and will revert back to its orginal state when the user is finished with the web application.


* Static website has stable content, every user will see and use the extact same page, a dynamic website content is pulled from a database of the fly, its content changes with the user such as ads appearing [Wix Blog Static vs Dynamic](https://developer.mozilla.org/en-US/docs/Web/HTTP/Session)


* Static webpages are made using HTML CSS JavaScript, are quick to make on the fly for impressive looking CV or for showing of math work in JupyterLite


* Dynamic while slower to make and load organizes data in a server-side like mongo or sql to dynamically change the website, giving the user the ablity to personalize the website.



## Setup GitHub Pages
***
* To use GitHub Pages you create a repository and enable Pages in the repository settings.


* You can specify a repository branch to be your website normally its main/master.


* Or you can use GitHub Actions, more below.

## Hugo
***


* Hugo is a staic site generator.


* Hugo can be written in markdown, JSON, etc. then can be published in HTML


* Requires a compilation step.


* Hugo is similar to GitHub pages in that it is a static site generator, with key features of Hugo being it takes <1 second to build a site page, supports a lot of content types without plugins, has many pre built templates, and you don't have to use markdown you can output in other formats such as JSON.


* [Hugo quick start](https://gohugo.io/getting-started/quick-start/)

## GitHub Actions
***

* GitHub Actions are scripts that run based on triggers related to a GitHub repository.


* Typically they are triggered when a commit is pushed to GitHub.


* The script might run a set of tests to ensure the code works, Or the script might build a website from the codebase.


* Technically Actions are config files, as opposed to scripts.
