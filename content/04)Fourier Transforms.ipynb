## Complex Numbers
***

Complex numbers are the sum of real & imaginary numbers, complex numbers can be written as a + ib where a is the real number and ib being the imaginary number, complex number are often used to find the square root of negative numbers, the concept of complex numbers dates back to the first century.

Real numbers refer to any number which isn't a complex number e.g 5, -5, 5.5, -5.5, 5/1 and so on.

Imaginary numbers are number that when it's squared it has a negative result, such imaginary numbers may also be complex numbers.

#### [Complex Numbers;](https://www.cuemath.com/numbers/complex-numbers/)Cuemath
#### [Imaginary Numbers;](https://www.livescience.com/42748-imaginary-numbers.html)livescience

```python
# Python code to demonstrate the working of complex, real and imaginary numbers
  
# cmath provides mathematical functions for complex numbers and can accept int, floating point numbers, or complex numbers
import cmath
  
# Initializing real numbers
x = 5
y = 3
  
# converting x and y into complex number, can be represented as x + yi
# the complex command does this by taking a real number(x) and a imaginary number(y)
z = complex(x,y);
  
# printing real and imaginary part of complex number
# end is used to print at the end of the output
print ("The real part of complex number is : ",end="")
print (z.real) # real represents a real number
  
print ("The imaginary part of complex number is : ",end="")
print (z.imag) # imag represents a imaginary number
```

```python
# Re-initializing real numbers
x = 2
y = 3j
  
# converting x and y into complex number, can be represented as x + yi
z = complex(x,y);
  
# printing phase of a complex number
# the phase method can return a phase of a complex number
# A complex number can be expressed in magnitude and angle
print ("The phase of complex number is : ",end="")
print (cmath.phase(z))
```

```python
# Re-initializing real numbers
x = 1.0
y = 1.0
  
# converting x and y into complex number, can be represented as x + yi
z = complex(x,y);
  
# converting complex number into polar
# polar method is used to covert a complex number into polar coordinates
w = cmath.polar(z)
  
# print
print ("The modulus and argument of polar complex number is : ",end="")
print (w)
  
# converting complex number into rectangular
# rect method used which returns the smallest rectangle that contains all non-empty rectangles a list argument
w = cmath.rect(1.4142135623730951, 0.7853981633974483)
  
# printing rectangular form of complex number
print ("The rectangular form of complex number is : ",end="")
print (w)
```

## Fourier Transforms
***
[SciPy Manual Fourier Transforms;](https://docs.scipy.org/doc/scipy/tutorial/fft.html#fast-fourier-transforms)


```python
# Used for visual effects such as graphs & plots
import matplotlib.pyplot as plt

# Numerical arrays.
import numpy as np

# Fourier Transforms
import scipy
```

```python
#imports for fft
from scipy.fft import fft, ifft, fftn, rfftn

#changed values 
x = np.array([0.5, 2.5, 1.0, -1.0, 1.5, 2.0, 3.0, -2.0])
```

```python
y = fft(x)
y
# in the below output numbers with j are complex numbers
```

```python
# ifft is the inverse of fft
yinv = ifft(y)
yinv
```

```python
# n is the number of points used to calculate fft
# mgrid is a function used to get a multi dimensional grid
xtwo = np.mgrid[:3, :3, :3][0]
z = fftn(xtwo)
z
```

```python
# np.random is a random sampling funtion
rng = np.random.default_rng()
# .pi returns value of PI
[X, Y] = np.meshgrid(2 * np.pi * np.arange(200) / 12,
                     2 * np.pi * np.arange(200) / 34)
```

```python
# sin function to calculate trigonometric sine, cos function to calculate trigonometric cosine
# uniform draws samples from a uniform distribution
S = np.sin(X) + np.cos(Y) + rng.uniform(0, 1, X.shape)
FS = scipy.fft.fftn(S)
# np.abs is function used to calculate absolute value in a given array
plt.imshow(np.log(np.abs(scipy.fft.fftshift(FS))**2))
```

```python
# computes one dimensional point discrete Fourier Transform (DFT) of a real-valued array
# np.ones returns a new array of shape and data type, where the elements value is set to 1
xthree = np.ones([2,2,2,2])
r = rfftn(xthree)
r
```
