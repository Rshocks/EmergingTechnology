## Complex Numbers
***

Complex numbers are the sum of real & imaginary numbers, complex numbers can be written as a + ib where a is the real number and ib being the imaginary number, complex number are often used to find the square root of negative numbers, the concept of complex numbers dates back to the first century.

Real numbers refer to any number which isn't a complex number e.g 5, -5, 5.5, -5.5, 5/1 and so on.

Imaginary numbers are number that when it's squared it has a negative result, such imaginary numbers may also be complex numbers.

#### [Complex Numbers;](https://www.cuemath.com/numbers/complex-numbers/)Cuemath
#### [Imaginary Numbers;](https://www.livescience.com/42748-imaginary-numbers.html)livescience

```python
# Python code to demonstrate the working of complex, real and imaginary numbers
  
# cmath provides mathematical functions for complex numbers and can accept int, floating point numbers, or complex numbers
import cmath
  
# Initializing real numbers
x = 5
y = 3
  
# converting x and y into complex number, can be represented as x + yi
# the complex command does this by taking a real number(x) and a imaginary number(y)
z = complex(x,y);
  
# printing real and imaginary part of complex number
# end is used to print at the end of the output
print ("The real part of complex number is : ",end="")
print (z.real) # real represents a real number
  
print ("The imaginary part of complex number is : ",end="")
print (z.imag) # imag represents a imaginary number
```

```python
# Re-initializing real numbers
x = 2
y = 3j
  
# converting x and y into complex number, can be represented as x + yi
z = complex(x,y);
  
# printing phase of a complex number
# the phase method can return a phase of a complex number
# A complex number can be expressed in magnitude and angle
print ("The phase of complex number is : ",end="")
print (cmath.phase(z))
```

```python
# Re-initializing real numbers
x = 1.0
y = 1.0
  
# converting x and y into complex number, can be represented as x + yi
z = complex(x,y);
  
# converting complex number into polar
# polar method is used to covert a complex number into polar coordinates
w = cmath.polar(z)
  
# print
print ("The modulus and argument of polar complex number is : ",end="")
print (w)
  
# converting complex number into rectangular
# rect method used which returns the smallest rectangle that contains all non-empty rectangles a list argument
w = cmath.rect(1.4142135623730951, 0.7853981633974483)
  
# printing rectangular form of complex number
print ("The rectangular form of complex number is : ",end="")
print (w)
```

## Quadratic Formula
***

The name quadratic comes from "quad" meaning square, below are two quadratic formulas

- x is the variable we have to figure out as it is the unknown


- The solution to a quadratic equation may be complex or real


- There is always guaranteed to be at least one solution


- If we use the above as an example, the values a, b, c represent the known numbers, x is the unknown number and the whole reason for such a formula is to find the value of x(squared) and x, if x is not squared then it is not a quadratic equation.

#### [Quadratic Equation;](https://www.mathsisfun.com/algebra/quadratic-equation.html)mathisfun
#### [Quadratic Equations for python;](https://www.geeksforgeeks.org/python-program-to-solve-quadratic-equation/)geeksforgeeks

```python
# formula of establishing the known numbers
def formula(a, b, c):
    # sqrt returns the square root of any number imported by math
    undersqrt = (b**2 - 4 * a * c)**0.5
    return [(-b - undersqrt) / (2 * a), (-b + undersqrt) / (2 * a)]
```

```python
# y = x^2 + 4 
formula(1, 0, 4)
```

```python
# import cmath is already done from above in complex numbers section
# declare three known values
x = 1
y = 5
z = 6

# calculate the discriminant
w = (y**2) - (4*x*z)

# find two solutions
sol1 = (-y-cmath.sqrt(w))/(2*x)
sol2 = (-y+cmath.sqrt(w))/(2*x)

# print solutions, format places the values inside {} for readability
print('The solution are {0} and {1}'.format(sol1,sol2))
```

```python
# Used for visual effects such as graphs & plots
import matplotlib.pyplot as plt

# Numerical arrays.
import numpy as np

# Fourier Transforms
import scipy
```

```python
#imports for fft
from scipy.fft import fft, ifft, fftn, rfftn

#changed values 
x = np.array([0.5, 2.5, 1.0, -1.0, 1.5, 2.0, 3.0, -2.0])
```

```python
y = fft(x)
y
# in the below output numbers with j are complex numbers
```

```python
# n is the number of points used to calculate fft
# mgrid is a function used to get a multi dimensional grid
xtwo = np.mgrid[:3, :3, :3][0]
z = fftn(xtwo)
z
```

```python
# np.random is a random sampling funtion
rng = np.random.default_rng()
# .pi returns value of PI
[X, Y] = np.meshgrid(2 * np.pi * np.arange(200) / 12,
                     2 * np.pi * np.arange(200) / 34)
```

```python
# sin function to calculate trigonometric sine, cos function to calculate trigonometric cosine
# uniform draws samples from a uniform distribution
S = np.sin(X) + np.cos(Y) + rng.uniform(0, 1, X.shape)
FS = scipy.fft.fftn(S)
# np.abs is function used to calculate absolute value in a given array
plt.imshow(np.log(np.abs(scipy.fft.fftshift(FS))**2))
```

```python
# computes one dimensional point discrete Fourier Transform (DFT) of a real-valued array
# np.ones returns a new array of shape and data type, where the elements value is set to 1
xthree = np.ones([2,2,2,2])
r = rfftn(xthree)
r
```

```python
#SciPy libary in python for computing 
#.fft used to compute fast fourier transformation
#.fftfreq used to return sample Discrete Fourier Transform sample frequencies
from scipy.fft import fft, fftfreq

# Number of sample points
N = 600
# sample spacing
T = 1.0 / 800.0

# linspace creates numeric sequences
x = np.linspace(0.0, N*T, N, endpoint=False)

# Trigonometric function for sine
y = np.sin(50.0 * 2.0*np.pi*x) + 0.5*np.sin(80.0 * 2.0*np.pi*x)

# using .ffy algorithm for fast Fourier transform
# .fftfreq returns Discrete Fourier Transform and its frequencies
yf = fft(y)
xf = fftfreq(N, T)[:N//2]

#Plot the above onto a graph
plt.plot(xf, 2.0/N * np.abs(yf[0:N//2]))
plt.grid()
plt.show()
```

```python
# blackman, returns a blackman window
from scipy.signal import blackman

# Number of sample points
N = 600

# sample spacing
T = 1.0 / 800.0

# linspace creates numeric sequences
x = np.linspace(0.0, N*T, N, endpoint=False)

# Trigonometric function for sine
y = np.sin(50.0 * 2.0*np.pi*x) + 0.5*np.sin(80.0 * 2.0*np.pi*x)
yf = fft(y)

# blackman, returns a blackman window
w = blackman(N)

# using .ffy algorithm for fast Fourier transform
# .fftfreq returns Discrete Fourier Transform and its frequencies
ywf = fft(y*w)
xf = fftfreq(N, T)[:N//2]

# Ploting
plt.semilogy(xf[1:N//2], 2.0/N * np.abs(yf[1:N//2]), '-b')
plt.semilogy(xf[1:N//2], 2.0/N * np.abs(ywf[1:N//2]), '-r')
plt.legend(['FFT', 'FFT w. window'])
plt.grid()
plt.show()
```

```python
# .ffshift shifts zero frequency to centre of the spectrum
from scipy.fft import fftshift

# number of signal points
N = 400

# sample spacing
T = 1.0 / 800.0

# linspace creates numeric sequences
x = np.linspace(0.0, N*T, N, endpoint=False)

# exp helps calculate the exponential elements of a given array
y = np.exp(50.0 * 1.j * 2.0*np.pi*x) + 0.5*np.exp(-80.0 * 1.j * 2.0*np.pi*x)

# using .ffy algorithm for fast Fourier transform
yf = fft(y)

# .fftfreq returns Discrete Fourier Transform and its frequencies
xf = fftfreq(N, T)

# .ffshift shifts zero frequency to centre of the spectrum
xf = fftshift(xf)
yplot = fftshift(yf)

# Plotting
plt.plot(xf, 1.0/N * np.abs(yplot))
plt.grid()
plt.show()
```

## Kronecker product
***
- In maths, kronecker product is a fucntion which of two matrices of random size which will outcome in a block matrix, it is oftern used to solve complex problems in linear algebra.


- Another description for the Kronecker product is an operation which forms two matrices into a combined matrix, this matrix contains all results of the two previous matrices

[Kronecker-product;](https://www.statlect.com/matrix-algebra/Kronecker-product)Statlect

[Kronecker-product Wiki;](https://en.wikipedia.org/wiki/Kronecker_product)Wikipedia

```python
# kron computes two Kronecker product
# eye returns a two-dimensional array with ones on the diagonal and zeros elsewhere
np.kron(np.eye(2), np.ones((2,2)))
```

```python
# .reshape allows to change the shape of an array 
# .arange is a array creation routine founded on numerical ranges
a = np.arange(100).reshape((2,5,2,5))
b = np.arange(24).reshape((2,3,4))

# kron computes two Kronecker product
c = np.kron(a,b)
c.shape
```

```python
# Assigning varibales
I = (1,3,0,2)
J = (0,2,1)
J1 = (0,) + J             # extend to ndim=4

# .shape used to get the current shape of a given array
S1 = (1,) + b.shape

# tuple array
K = tuple(np.array(I) * np.array(S1) + np.array(J1))
c[K] == a[I]*b[J]
```
