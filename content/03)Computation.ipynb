### Big O Notation
***
Big O details the complexity of code we use in the standards of algebra. It is a system for measuring the growth of an algorithm, it doesn't measure speed it measures the number of operations it takes to complete

### Exercise 1

```python
# Stylesheets defined in Matplotlib
plt.style.use('bmh')

# Set up runtime
n = np.linspace(1, 10, 1000)
#labels
labels = ['Constant', 'Linear', 'Log Linear', 'Quadratic', 'Exponential']
# types of big o
big_o = [np.ones(n.shape), n, n * np.log(n), n**2, 2**n]

# Plot setup
plt.figure(figsize=(8, 4))
plt.ylim(0, 50)

for i in range(len(big_o)):
    plt.plot(n, big_o[i], label=labels[i])

plt.legend(loc=0)
```

- Linear O(n) an example of linear is a search in an array to which every element in said array is being checked against the query each time, rate of growth would still be on the faster end of the scales.
<br></br>
- Log Linear O(n * log n) is a linear time complexity multiplied by a logarithmic time complexity example of such is merge sort which accepts two arrays then shifts them into a new array, to sort we need one iteration over an element O(n) then divide our input into two arrays O(log n) merge them together to get log linear.

### Turing Machine
***

The **Turing Machine** is a theory based machine which can control symbols which appear on a tape strip, based on a set of guidelines.

#### [On Computable Numbers, with an application to the Entscheidungsproblem;](https://www.cs.virginia.edu/~robins/Turing_Paper_1936.pdf)A. M. Turing; Nov 1937; Proceedings of the London Mathematical Society.

Below simulates a turning machine

```python
# State table.
# tuple can't change
table = {
    ('X', '_'): ['_', 'R', 'T'],
    ('X', '0'): ['0', 'R', 'X'],
    ('X', '1'): ['1', 'R', 'Y'],
    ('Y', '_'): ['_', 'R', 'F'],
    ('Y', '0'): ['0', 'R', 'Y'],
    ('Y', '1'): ['1', 'R', 'X'],
}

# Tape input. accepts strings that end in even number, doesnt accept uneven numbers
tape = list('0101111')
# Position on tape.
pos = 0
# Initial state is first in table.
state = 'X'

# Keep going while we are not in a halting state.
while state not in ['T', 'F']:
    # Print the current status.
    print(''.join(tape[:pos]) + state + ''.join(tape[pos:]))
    # Get the row of the table.
    row = table[(state, tape[pos])]
    # Overwrite the symbol.
    tape[pos] = row[0]
    # Move left or right.
    if row[1] == 'R':
        # Put blanks on tape as necessary.
        if pos == len(tape) - 1:
            tape = tape + ['_']
        # Increase position.
        pos = pos + 1
    else:
        # Put blanks on tape as necessary.
        if pos == 0:
            tape = ['_'] + tape
            # The position on the tape has to move with it.
            pos = pos + 1
        
        # Increase position.
        pos = pos - 1
    # Update the state.
    state = row[2]

# Print the current status.
print(''.join(tape[:pos]) + state + ''.join(tape[pos:]))
```

### Bitwise Operations
***

There are a host of programming languages have operators which work at the bit level such as C or C++.

Bitwise is an operation which works with single bits the smallest unit of data in computation, most langauges will change these bits and represent them in groups we might be more familiar with such as 8 16 32 64 and these multiples can be known as byte arrays.

#### [A Definition of bitwise](https://www.techtarget.com/whatis/definition/bitwise)

```python
# Two integers, filled using binary notation.
x = 0b0011
y = 0b0101
```

```python
# Print x in binary with at least 8 0's and 1's, left padding with 0's if needed.
print(f'{x:08b}')
```

### Bitwise Operations in Python
***

#### [Bitwise in Python](https://realpython.com/python-bitwise-operators/)
1. Python lets you manipulate bits of data down to the finest grain
2. These bitwise operators can and are used in algorithms, or error detection and in controlling physical devices 

- & = Bitwise AND
- | = Bitwise OR
- ^ = Bitwise XOR (exclusive OR)
- ~ = Bitwise NOT
- << = Bitwise left shift, Bitwise right shift being the opposite

```python
# and

# Print the & truth table.
print(f'  {x:04b}')
print(f'& {y:04b}')
print(f'------')
print(f'  {x&y:04b}')
```

```python
# or

# Print the | truth table.
print(f'  {x:04b}')
print(f'| {y:04b}')
print(f'------')
print(f'  {x|y:04b}')
```

```python
# xor

# Print the ^ truth table.
print(f'  {x:04b}')
print(f'^ {y:04b}')
print(f'------')
print(f'  {x^y:04b}')
```

```python
# not

# Print the ~ truth table.
# Note we have to be careful here because x is a signed integer.
print(f'~ {0b01011:04b}')
print(f'------')
print(f'  {~0b01011:04b}')
```

```python
# not again

# Print the ~ truth table.
# Note we have to be careful here because x is a signed integer.
print(f'~ {x:04b}')
print(f'------')
print(f'  {0b1111&(~x):04b}')
```

```python
# Left shift.
print(f'{x:06b} << 2 = {x << 2:06b}')
```

```python
# Right shift.
print(f'{y:06b} >> 2 = {y >> 2:06b}')
```
